```python
## 사전작업
## 임시/취합 Dataset 정의
df_data_cal = pd.DataFrame()

## 취합할 데이터의 "locdate"를 datetime으로 변경
df_dateRange_d['DATETIME'] = pd.to_datetime(df_dateRange_d['DATETIME'], format='%Y%m%d') #, format='%Y%m%d %H%M%S')
df_specDay['DATETIME'] = pd.to_datetime(df_specDay['locdate'], format='%Y%m%d') #, format='%Y%m%d %H%M%S')
df_restDay['DATETIME'] = pd.to_datetime(df_restDay['locdate'], format='%Y%m%d') #, format='%Y%m%d %H%M%S')

## Column 정렬 및 삭제
df_specDay = df_specDay[['DATETIME', 'dateKind', 'dateName', 'isHoliday']]
df_restDay = df_restDay[['DATETIME', 'dateKind', 'dateName', 'isHoliday']]

df_dateRange_d
df_specDay
df_restDay
```

```python
df_restDay
```

```python
df_data_cal = pd.merge(df_dateRange_d, df_specDay, how='left', on=['DATETIME'])
df_data_cal = df_data_cal.drop_duplicates()
df_data_cal.head(5)

df_data_cal = pd.merge(df_data_cal, df_restDay, how='left', on=['DATETIME', 'dateKind', 'dateName', 'isHoliday'])
df_data_cal = df_data_cal.drop_duplicates()
df_data_cal.head(5)
```

```python
## Mod Column
df_data_cal = df_data_cal.rename(columns={'isHoliday':'rest_YN', 'dateName':'name_of_holiday'})

## 불필요 Column(seq, remark) 제외 및 순서 변경
df_data_cal['DATETIME'] = pd.to_datetime(df_data_cal['DATETIME'])
df_data_cal['WEEK'] = df_data_cal['DATETIME'].dt.strftime('%G-%V') 

df_data_cal = df_data_cal[['DATETIME', 'YEAR', 'MONTH', 'DAY', 'WEEK'
                           , 'day_of_the_week', 'code_day_of_the_week', 'rest_YN', 'name_of_holiday'
                           , 'HOUR', 'MINUTE']]

print(df_data_cal.info())
df_data_cal.head(10)
```

#### 01-03-02. 휴일여부 체크 및 Numeric 변환

```python
## 휴일 여부
## Y -> 1 / N -> 0으로 변환
df_data_cal.loc[(df_data_cal['rest_YN']=='Y'), 'rest_YN'] = '1'
df_data_cal.loc[(df_data_cal['rest_YN']=='N'), 'rest_YN'] = '0'
df_data_cal['rest_YN'] = df_data_cal['rest_YN'].fillna('0')

## 휴일 명칭
## 없을 경우 '-'로 변경
df_data_cal['name_of_holiday'] = df_data_cal['name_of_holiday'].fillna('-')

## 주말인 경우 휴일로 지정
## Sunday / Saturday는 휴일 처리
df_data_cal.loc[(df_data_cal['day_of_the_week']=='Sunday') 
                | (df_data_cal['day_of_the_week']=='Saturday'), 'rest_YN'] = '1'
```
